```python
import matplotlib.pyplot as plt
import csv
```

```python
def generate_bar_chart(labels, values):
  fig, ax = plt.subplots()
  ax.bar(labels, values)
  plt.show()
  
def generate_pie_chart(labels, values):
  fig, ax = plt.subplots()
  ax.pie(values, labels=labels)
  ax.axis('equal')
  plt.show()
```

```python
def get_population_history(country):
    
    with open('../csv/poblacion.csv', 'r') as csvFile:
        reader = csv.DictReader(csvFile)
    
        data =list(filter(lambda row: row['Country'] == country, reader))
        
        if len(data) == 0:
            raise ValueError('El pais ingresado no existe en la base de datos')
        
        data_country = data[0]
        
        return {
            '2022': float(data_country['2022 Population']),
            '2020': float(data_country['2020 Population']),
            '2015': float(data_country['2015 Population']),
            '2010': float(data_country['2010 Population']),
            '2000': float(data_country['2000 Population']),
            '1990': float(data_country['1990 Population']),
            '1980': float(data_country['1980 Population']),
            '1970': float(data_country['1970 Population'])
        }
```

```python
data = get_population_history('Argentina')
print(data)
years = data.keys()
valuesYears = data.values()
generate_bar_chart(years, valuesYears)
```

```python
def get_population_percentage (Continent=False):
    dict_contries = {}
    
    with open('../csv/poblacion.csv', 'r') as csvFile:
        reader = csv.DictReader(csvFile)
        
        if Continent:
            reader = list(filter(lambda row: row['Continent'] == Continent,reader))
        
        for row in reader:
            dict_contries[row['Country']] = row['World Population Percentage']
            
    return dict_contries
```

```python
data = get_population_percentage('Europe')
contries = data.keys()
percentage = data.values()

generate_pie_chart(contries, percentage)
```
